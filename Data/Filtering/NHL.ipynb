```python
# Import Dependencies
import pandas as pd
import numpy as np
#from sklearn.preprocessing import LabelEncoder
import sqlalchemy
from sqlalchemy import create_engine
import psycopg2
#postgresql://user:password@localhost:5432/databasename
engine = sqlalchemy.create_engine('postgresql://postgres:Lolhehe!0103@localhost:5432/Athletes_Career_Lifetimes')

# Import File
nhl_df = pd.read_csv("../../Data/DataSet/NHL.csv", sep=",", encoding='latin-1')
nhl_df
```

```python
# Keep Players who played after 2004
nhl_df = nhl_df[(nhl_df["Season"] > 2004)]
nhl_df
```

```python
# Keep Players who played before 2018
nhl_df = nhl_df[(nhl_df["Season"] < 2018)]
nhl_df
```

```python
# Filter necessary columns
nhl_df = nhl_df[['Player','sport', 'Age', 'Pos', 'Season']]
nhl_df
```

```python
# Filter players to youngest age starting at 2005
min_age_df = nhl_df.groupby(["Player"])["sport","Age", "Pos", "Season"].min().reset_index()
min_age_df
```

```python
# Filter players to oldest age starting ending at 2017
max_age_df = nhl_df.groupby(["Player"])["sport","Age", "Pos", "Season"].max().reset_index()
max_age_df
```

```python
# Assign min df to start df
start_nhl_df = min_age_df

start_nhl_df["Years_Played"] = (max_age_df["Age"] - min_age_df["Age"]) + 1

# Rename columns
start_nhl_df = start_nhl_df.rename(columns={'Player': 'playerID', 'Age': 'start_age', 'Pos': 'position', 'Season': 'start_year', 'Years_Played': 'years_played'})

# Reorder columns
start_nhl_df = start_nhl_df[['playerID','sport', 'position', 'start_year', 'start_age', 'years_played']]

# Display Data Frame
start_nhl_df
```

```python
# Get rid of multiple positions
start_nhl_df = start_nhl_df[start_nhl_df.position != 'LW/C']
start_nhl_df = start_nhl_df[start_nhl_df.position != 'LW/RW']
start_nhl_df['position'].describe()
```
