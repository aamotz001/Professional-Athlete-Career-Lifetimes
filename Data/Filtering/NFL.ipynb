```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
#postgresql://user:password@localhost:5432/databasename
engine = sqlalchemy.create_engine('postgresql://postgres:gonoles@localhost:5432/Athletes_Career_Lifetimes')
```

```python
file = "../../Data/DataSet/NFL.csv"
nfl_df = pd.read_csv(file)
nfl_df.head()
```

```python
#finding the retirement age of listed players - subtracting the end of career year with the player's birth year
nfl_df['birth_year'] = pd.DatetimeIndex(nfl_df['birth_date']).year
nfl_df.head()
```

```python
#create column for retirement age
nfl_df['retire_age'] = nfl_df['year_end'] - nfl_df['birth_year']
nfl_df.head()
```

```python
#filter retirement ages for those who started career before 1970
nfl_df = nfl_df[nfl_df['year_start'] >= 1970]
nfl_df.head(30)
```

```python
#filter retirement ages for those who were born before 1970
nfl_df = nfl_df[nfl_df['birth_year'] >= 1970]
nfl_df.head()
```

```python
#filter retirement ages for incorrect data
nfl_df = nfl_df[nfl_df['retire_age'] >= 18]
nfl_df.head()
```

```python
#find career length/years played and starting age
nfl_df['years_played'] = nfl_df['year_end'] - nfl_df['year_start'] +1
nfl_df['start_age'] = nfl_df['year_start'] - nfl_df['birth_year']
nfl_df.head(30)
```

```python
nfl_df = nfl_df.drop(columns=['first_name', 'last_name', 'birth_year', 'birth_city', 'birth_state', 'birth_country', 'birth_date', 'college', 'draft_team', 'draft_round', 'draft_pick', 'draft_year', 'height', 'weight', 'death_date', 'death_city', 'death_state', 'death_country'])
nfl_df.head(30)
```

```python
#next -> filter down to 5 positions (rb, qb, wr, lb, OT)
position_list = ['RB', 'QB', 'WR', 'LB', 'CB']
retired_nfl_df = nfl_df[nfl_df['position'].isin(position_list)]
retired_nfl_df.head(30)
```

```python
retired_nfl_df = retired_nfl_df.rename(columns={'name':'playerid','year_start':'start_year', 'year_end':'retire_year'})
retired_nfl_df.head()
```

```python
retired_nfl_df['retire_age'] = retired_nfl_df['retire_age'].map("{:.0f}".format)
retired_nfl_df['start_age'] = retired_nfl_df['start_age'].map("{:.0f}".format)
```

```python
starting_nfl_df = retired_nfl_df[['playerid', 'sport', 'position', 'start_year', 'start_age', 'years_played']].copy()
retiring_nfl_df = retired_nfl_df[['playerid','sport', 'position', 'retire_year', 'retire_age', 'years_played']].copy()
starting_nfl_df.head()
```

```python
retiring_nfl_df = retiring_nfl_df.drop_duplicates(subset=['playerid'])
starting_nfl_df = starting_nfl_df.drop_duplicates(subset=['playerid'])
```

```python

starting_nfl_df['start_age'] = starting_nfl_df['start_age'].astype('int64')
retiring_nfl_df['retire_age'] = retiring_nfl_df['retire_age'].astype('int64')
```

```python
#use if_exists=replace parameter to replace data if table already exists

starting_nfl_df.to_sql(
    name='nfl_starting',
    con=engine,
    index=False,
    if_exists='replace'
)

retiring_nfl_df.to_sql(
    name='nfl_retiring',
    con=engine,
    index=False,
    if_exists='replace'
)
```
