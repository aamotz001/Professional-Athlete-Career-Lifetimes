```python
# Import Dependencies
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Import File
nhl_df = pd.read_csv("../../Data/DataSet/NHL.csv", sep=",", encoding='latin-1')
nhl_df
```

```python
# Keep Players who played after 2004
nhl_df = nhl_df[(nhl_df["Season"] > 2004)]
nhl_df
```

```python
# Keep Players who played before 2018
nhl_df = nhl_df[(nhl_df["Season"] < 2018)]
nhl_df
```

```python
# Filter necessary columns
nhl_df = nhl_df[['Player', 'Age', 'Pos', 'Season']]
nhl_df
```

```python
# Filter players to youngest age starting at 2005
min_age_df = nhl_df.groupby(["Player"])["Age", "Pos", "Season"].min().reset_index()
min_age_df
```

```python
# Filter players to oldest age starting ending at 2017
max_age_df = nhl_df.groupby(["Player"])["Age", "Pos", "Season"].max().reset_index()
max_age_df
```

```python
# Assign min df to start df
start_nhl_df = min_age_df

start_nhl_df["Years_Played"] = (max_age_df["Age"] - min_age_df["Age"]) + 1

# Rename columns
start_nhl_df = start_nhl_df.rename(columns={'Player': 'player', 'Age': 'start_age', 'Pos': 'position', 'Season': 'start_year', 'Years_Played': 'years_played'})

# Reorder columns
start_nhl_df = start_nhl_df[['player', 'position', 'start_year', 'start_age', 'years_played']]

# Display Data Frame
start_nhl_df
```

```python
# Get rid of multiple positions
start_nhl_df = start_nhl_df[start_nhl_df.position != 'LW/C']
start_nhl_df = start_nhl_df[start_nhl_df.position != 'LW/RW']
start_nhl_df['position'].describe()
```

```python
start_nhl_df['position'].unique()
```

```python
lbl_encode = LabelEncoder()
lbl_encode.fit_transform(start_nhl_df['position'])
```

```python
start_nhl_df['positions'] = lbl_encode.fit_transform(start_nhl_df['position'])
# NHL position changed to numerical form
start_nhl_df['positions'].value_counts()
```

```python
add_columns = pd.get_dummies(start_nhl_df['position'])
add_columns
```

```python
start_nhl_df.join(add_columns)
```

```python
start_nhl_df.drop(['positions'], axis = 1, inplace = True)
start_nhl_df = start_nhl_df.join(add_columns)

# Reorder columns
start_nhl_df = start_nhl_df[['player','position', 'C', 'D', 'LW', 'RW', 'start_year', 'start_age', 'years_played']]
start_nhl_df
```

```python
# Export as CSV
start_nhl_df.to_csv("../../Data/Clean_Data/clean_start_nhl.csv", index=False)
```

```python
# Assign max df to end df
end_nhl_df = max_age_df

end_nhl_df["Years_Played"] = (max_age_df["Age"] - min_age_df["Age"]) + 1

# Rename columns
end_nhl_df = end_nhl_df.rename(columns={'Player': 'player', 'Age': 'retire_age', 'Pos': 'position', 'Season': 'retire_year', 'Years_Played': 'years_played'})

# Reorder columns
end_nhl_df = end_nhl_df[['player', 'position', 'retire_year', 'retire_age', 'years_played']]

# Display Data Frame
end_nhl_df
```

```python
# Get rid of multiple positions
end_nhl_df = end_nhl_df[end_nhl_df.position != 'C/LW']
end_nhl_df = end_nhl_df[end_nhl_df.position != 'RW/LW']
end_nhl_df = end_nhl_df[end_nhl_df.position != 'LW/C']
end_nhl_df['position'].describe()
```

```python
end_nhl_df['position'].unique()
```

```python
lbl_encode = LabelEncoder()
lbl_encode.fit_transform(end_nhl_df['position'])
```

```python
end_nhl_df['positions'] = lbl_encode.fit_transform(end_nhl_df['position'])
# NHL position changed to numerical form
end_nhl_df['positions'].value_counts()
```

```python
add_columns = pd.get_dummies(end_nhl_df['position'])
add_columns
```

```python
end_nhl_df.join(add_columns)
```

```python
end_nhl_df.drop(['positions'], axis = 1, inplace = True)
end_nhl_df = end_nhl_df.join(add_columns)

# Reorder columns
end_nhl_df = end_nhl_df[['player', 'position', 'C', 'D', 'LW', 'RW', 'retire_year', 'retire_age', 'years_played']]
end_nhl_df
```

```python
# Export as CSV
end_nhl_df.to_csv("../../Data/Clean_Data/clean_retire_nhl.csv", index=False)
```
