```python
# Import Dependencies
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout

# Export as CSV
nhl_start_df = pd.read_csv("../Data/Clean_Data/clean_start_nhl.csv", sep=",", encoding='latin-1')
nhl_start_df
```

```python
del nhl_start_df['player']
del nhl_start_df['position']
```

```python
nhl_start_df.head()
```

```python
nhl_start_df = nhl_start_df.astype('float32')
```

```python
nhl_start_df.head()
```

```python
nhl_start_df = nhl_start_df.to_numpy()
```

```python
# Print shape of numpy array
nhl_start_df.shape
```

```python
# Get training and testing dataset from array
training = nhl_start_df[0:1905]
testing = nhl_start_df[1905:]
```

```python
# Print training and testing dataset shape
print(training.shape)
print(testing.shape)
```

```python
# Separate true labels from training and testing datasets
training_features = training[:,0:-1]
training_labels = training[:,-1]
testing_features = testing[:,0:-1]
testing_labels = testing[:,-1]
```

```python
# Print shape of input that training_features and true labels which is training_labels
print(training_features.shape)
print(training_labels.shape)
```

```python
# Checking input shape to neural network
training_features[0].shape
```

```python
# Model definition
nn_model = Sequential()
nn_model.add(Input(shape=(6,)))

# Add our first Dense layer
nn_model.add(Dense(16, activation="relu"))
# Dropout Layer
nn_model.add(Dropout(0.2))

# Add our second Dense layer
nn_model.add(Dense(8, activation="relu"))
# Dropout Layer
nn_model.add(Dropout(0.2))

# Add last layer
nn_model.add(Dense(1, activation="sigmoid"))
```

```python
# Configure model training
nn_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# Train model
nn_model.fit(training_features, training_labels, epochs=100, validation_data=(testing_features, testing_labels))
```
